```python
def count_per_cat(col):
    vc = df[col].value_counts()  # Calculate value counts
    plt.figure(figsize=(12, 5))  # Set figure size

    # Bar plot
    plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
    ax = sns.countplot(x=df[col])
    ax.bar_label(ax.containers[0])  # Add labels to bars
    plt.title(f'Count plot for {col}')

    # Pie chart
    plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
    plt.pie(vc.values, labels=vc.index, autopct='%.2f%%')
    plt.title(f'Percentage distribution for {col}')

    plt.show()  # Display plots
    return vc  # Return value counts
    
```

```python
#with sorting
def count_per_cat(col):
    vc = df[col].value_counts()  # Calculate value counts
    plt.figure(figsize=(12, 5))  # Set figure size

    # Bar plot
    plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
    ax = sns.countplot(x=df[col],order=df[col].value_counts().sort_values(ascending=False).index)
    ax.bar_label(ax.containers[0])  # Add labels to bars
    plt.title(f'Count plot for {col}')

    # Pie chart
    plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
    plt.pie(vc.values, labels=vc.index, autopct='%.2f%%')
    plt.title(f'Percentage distribution for {col}')

    plt.show()  # Display plots
    return vc  # Return value counts
```

```python
cat_kpi = df.groupby(["pizza_category"]).agg(total_sales=("total_price","sum"),
                                              avg_sales = ("total_price","mean"),
                                              avg_unit_price   = ("unit_price","mean")).reset_index()
cat_kpi
```

```python
def bivar_fun1(col):
    w = df.groupby([col]).agg(total_sales=("total_price","sum"),
                                              avg_sales = ("total_price","mean"),
                                              avg_unit_price   = ("unit_price","mean")).reset_index()
    plt.figure(figsize=(12,10))
    plt.subplot(2,2,1)
    ax1=sns.barplot(data=w,x="pizza_category",y="total_sales",color="maroon")
    ax1.bar_label(ax1.containers[0])
    plt.title(f"barplot for {col} wrt total_sales")

    plt.subplot(2,2,2)
    ax2=sns.barplot(data=w,x="pizza_category",y="avg_sales",color="tan")
    ax2.bar_label(ax2.containers[0])
    plt.title(f"barplot for {col} wrt avg_sales")

    plt.subplot(2,2,3)
    ax3=sns.barplot(data=w,x="pizza_category",y="avg_sales",color="green")
    ax3.bar_label(ax3.containers[0])
    plt.title(f"barplot for {col} wrt avg_unit_price")

    plt.show()
    
```

```python
plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
ax1=sns.barplot(data=cat_kpi,x="pizza_category",y="total_sales",color="maroon")
ax1.bar_label(ax1.containers[0])
plt.title("barplot for category wrt total_sales")

plt.subplot(2,2,2)
ax2=sns.barplot(data=cat_kpi,x="pizza_category",y="avg_sales",color="tan")
ax2.bar_label(ax2.containers[0])
plt.title("barplot for category wrt avg_sales")

plt.subplot(2,2,3)
ax3=sns.barplot(data=cat_kpi,x="pizza_category",y="avg_sales",color="green")
ax3.bar_label(ax3.containers[0])
plt.title("barplot for category wrt total_sales")

plt.show()

```
