```python
description = pd.DataFrame(index=['observations(rows)', 'percent missing', 'dtype', 'range']) # creating a dataframe with custom indices
```

```python
# Ye code ek Pandas DataFrame ibm_df ke columns ko analyze kar raha hai aur har ek column ka summary statistics generate kar raha hai.
numerical = []
categorical = []
for col in ibm_df.columns:
    obs = ibm_df[col].size  # Total rows (1470)
    p_nan = round(ibm_df[col].isna().sum()/obs, 2)  # % missing
    # {} ke bina Python samajh nahi payega ki ye evaluate karna hai ya sirf as text likhna hai.
    num_nan = f'{p_nan}% ({ibm_df[col].isna().sum()}/{obs})'  #  in string Format missing
    dtype = 'categorical' if ibm_df[col].dtype == object else 'numerical'  # Text or number?
    numerical.append(col) if dtype == 'numerical' else categorical.append(col)  # Sort columns
    rng = f'{len(ibm_df[col].unique())} labels' if dtype == 'categorical' else f'{ibm_df[col].min()}-{ibm_df[col].max()}'  # Range
    description[col] = [obs, num_nan, dtype, rng]
numerical.remove('EmployeeCount')  # Drop constant
numerical.remove('StandardHours')  # Drop constant
pd.set_option('display.max_columns', 100) # displaying 100 columns atmost

display(description)
display(ibm_df.head())
```

```python
# defining a method for calculating accuracy,search time, hyperparameters etc for different models
def org_results(trials, hyperparams, model_name):
    fit_idx = -1
    for idx, fit  in enumerate(trials):
        hyp = fit['misc']['vals']
        xgb_hyp = {key:[val] for key, val in hyperparams.items()}
        if hyp == xgb_hyp:
            fit_idx = idx
            break
            
    train_time = str(trials[-1]['refresh_time'] - trials[0]['book_time'])
    acc = round(trials[fit_idx]['result']['accuracy'], 3)
    train_auc = round(trials[fit_idx]['result']['train auc'], 3)
    test_auc = round(trials[fit_idx]['result']['test auc'], 3)

    results = {
        'model': model_name,
        'parameter search time': train_time,
        'accuracy': acc,
        'test auc score': test_auc,
        'training auc score': train_auc,
        'parameters': hyperparams
    }
    return results
```
