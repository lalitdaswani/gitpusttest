```python
#left skewed check tail left side
#The mean is calculated by adding up all the values in the dataset and dividing by the number of values.
#In a left-skewed distribution, there are some smaller values (the left tail) that "pull" the mean toward the lower end.
#These outliers have a greater influence on the mean because it is sensitive to all data points, especially extreme ones.
#As a result, the mean ends up being lower than the median and mode.
#negative skweness means we have lower side outliers
```
