```python
import numpy as np 
import pandas as pd
```

```python
np.random.seed(20)
df=pd.DataFrame(np.random.randn(10,4),index=pd.date_range("2021-7-18",periods=10),columns=["a","b","c","d"])
df
```

```python
df["new"]=df["c"]+df["d"]
```

```python
df
```

```python
df["new_col_sqr_d"]= df["d"].apply(lambda x : x**2)
```

```python
df
```

```python
df["new_col_sqr_d_log"]= df["d"].apply(lambda x : np.log(x))
```

```python
df
```

```python

def test(x):
    if (x<0):
        return x
    else:
        return np.log(x)
```

```python
df["new_col_sqr_d_log1"]= df['d'].apply(test)
```

```python
df
```

```python
def test1(x):
    if x <0:
        return 1
    else:
        return 0
        
```

```python
df["new_a"] =df["a"].apply(test1)
```
